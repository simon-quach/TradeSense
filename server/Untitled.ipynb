```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
stock = yf.Ticker("NVDA")
```

```python
stock.info
```

```python
history = stock.history(period="1y")
```

```python
history
```

```python
# If you want to convert the dates to strings in the array
dates = history.index.strftime('%Y-%m-%d').to_numpy()
prices = history["Close"].to_numpy()
print(dates)
print(prices)
```

```python
# Assuming 'dates' and 'prices' are already defined as NumPy arrays
combined_data = [{'date': date, 'price': price} for date, price in zip(dates, prices)]

# Now, 'combined_data' is a list of dictionaries in the desired format
print(combined_data)
```
